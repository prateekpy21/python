```python
import numpy as np
from flair.embeddings import WordEmbeddings
from flair.embeddings import TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```python
glove_embeddings = WordEmbeddings('glove') #context independent
```

```python
sentence_1 = Sentence('apple released iphone 12 pro max in 2020')
```

```python
glove_embeddings.embed(sentence_1)
```

```python
for token in sentence_1:
    print(token)
    print(token.embedding)
    print("\n")
```

```python
sentence_2 = Sentence('an apple a day keeps the doctor away')
```

```python
glove_embeddings.embed(sentence_2)
```

```python
for token in sentence_2:
    print(token)
    print(token.embedding)
    print("\n")
```

```python
sentence_2[1]
```

```python
sentence_2[1].embedding
```

```python
sentence_2[1].embedding.shape
```

```python
glove_dist = distance.euclidean(np.array(sentence_1[0].embedding),
                                np.array(sentence_2[1].embedding))
```

```python
print("Distance between apple embeddings for Glove = {}".format(glove_dist))
```

```python
bert_embedding = TransformerWordEmbeddings('bert-base-multilingual-cased')
```
